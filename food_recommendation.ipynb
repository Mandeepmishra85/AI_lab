```python
food = food[['names','ingredients','diet','flavor_profile','course']]
```

```python
food['flavor_profile']=food['flavor_profile'].replace(
    '-1' , random.choice(['spicy','sweet']))
```

```python
food
```

```python
food['flavor_profile'].value_counts()
```

```python
food.isnull()
```

```python
def Convert(string):
    li = (string.split(","))
    return li

food['ingredients']=food['ingredients'].apply(Convert)
food['ingredients']
```

```python
food['ingredients']=food['ingredients'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
food
```

```python
food['diet']=food['diet'].apply(Convert)
food['flavor_profile']=food['flavor_profile'].apply(Convert)
food['course']=food['course'].apply(Convert)
```

```python
food
```

```python
food['food_name'] = food['names']
food['tags']=food['ingredients']+food['diet']+food['flavor_profile']+food['course']
```

```python
food
```

```python
new_df=food[['food_name','tags']]
new_df['tags'] =new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
def Convert(string):
    li = string.lower()
    return li
```

```python
new_df['food_name']=new_df['food_name'].str.lower()
```

```python
new_df
```

```python

#####Imporint feature extractor################

import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new_df['tags']).toarray()


print(vector)
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
  y =[]

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)


new_df['tags']=new_df['tags'].apply(stem)
print(new_df)
similarity = cosine_similarity(vector)
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[1:6]
similarity[0]
```

```python
######OUTPUT####
def recommend(food):
    anime_index = new_df[new_df['food_name'] == food].index[0]
    distances = similarity[anime_index]
    food = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    for i in food:
        print(new_df.iloc[i[0]].food_name)
recommend('egg chowmein')
```
